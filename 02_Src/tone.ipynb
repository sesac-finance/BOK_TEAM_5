```python
def token_ngram(text):
    from ekonlpy.sentiment import MPCK
    mpck = MPCK()

    token = mpck.tokenize(text)
    ngram = mpck.ngramize(token)
        
    return ngram
```

```python
df['ngrams']= df['content'].apply(lambda x: token_ngram(str(x)))
```

```python
hakish_dict = pd.read_csv('../../02_Src/hakish_dict.csv')
dovish_dict = pd.read_csv('../../02_Src/dovish_dict.csv')
```

```python
hakish_dict = hakish_dict.rename(columns={'Unnamed: 0':'ngrams'})
dovish_dict = dovish_dict.rename(columns={'Unnamed: 0':'ngrams'})
```

```python
hakish_ls = hakish_dict['ngrams'].values.tolist()
dovish_ls = dovish_dict['ngrams'].values.tolist()
```

```python
def tone_c(haw, dov):
    try:
        tone = (haw - dov) / (haw + dov)
    except:
        tone = 0
    
    return tone 
```

```python
def cnt_haw_dov(ngram):
    haw, dov = 0, 0
    
    for n in ngram:
        if n in hakish_ls:
            haw+=1
        elif n in dovish_ls:
            dov+=1
                    
    tone = tone_c(haw, dov)
    return tone
```

```python
proceedings_rate['tone'] = proceedings_rate['ngrams'].apply(lambda x: cnt_haw_dov(x))
```
