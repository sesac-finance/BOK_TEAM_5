```python
import requests
from bs4 import BeautifulSoup
import os
```

```python
os.chdir('../03_Data/의사록')
```

```python
def download(download_url, file_name):
    with open(file_name, "wb") as file:  
        response = requests.get(download_url)
        file.write(response.content)  
```

```python
pages =35
for page in range(1,pages+1):
    url = 'https://www.bok.or.kr/portal/bbs/B0000245/list.do?menuNo=200761&pageIndex=' + str(page)
    res = requests.get(url)
    soup = BeautifulSoup(res.text, 'html.parser')
    results = soup.select('a[href*="&fileSn="]')
    title_results = soup.select('.titlesub')

    for i in range(20):
        try:
            if results[i].attrs['title'][-3:] =='pdf':
                download_url = 'https://www.bok.or.kr' + results[i].attrs['href']
                # print(download_url)
                print(i)
        except:
            pass

        if i%2 == 1:
            title = title_results[i//2].text.rstrip().split('(')[2][:-1]
            if title[-1:] == '.':
                title = title + 'pdf'
            else:
                title = title + '.pdf'
            download(download_url,title)
            print(download_url)
            print(title)
        
    if title[:4] == '2010':
        break
```
