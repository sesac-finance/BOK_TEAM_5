```python
import pandas as pd
```

```python
data = pd.read_excel('../BOK/한국금리.xlsx')
```

```python
data['월'] = data['변경월일'].apply(lambda x : x[:2] )
data['일'] = data['변경월일'].apply(lambda x : x[3:6] )
```

```python
data['연월일'] = data.apply(lambda x : str(x.변경년) +'-' + x.변경월일[:2] + '-' + x.변경월일[4:6],axis=1 )
```
